```python
import numpy as np
import matplotlib.pyplot as plt
def Randomwalk1D(n): #n here is the no. of steps that we require
   x = 0
   y = 0
   xposition = [0] #starting from origin (0,0)
   yposition = [0] 
   for i in range (1,n+1):
       step = np.random.uniform(0,1)
       if step > 0.5: # if step is less than 0.5 we move right    
           x += 1  #moving right in x direction
           y += 1 
       if step < 0.5: # if step is greater than 0.5 we move left  
           x += -1  #moving left in x direction
           y += 1 
 
       xposition.append(x)
       yposition.append(y)
   return [xposition,yposition]
Randwalk = Randomwalk1D(50) #creating an object for the Randomwalk1D class and passing value of n as 50
plt.plot(Randwalk[0],Randwalk[1],'r-')
plt.title("Random Walk on Integers")
plt.grid()
plt.xlabel('State')
plt.ylabel('No. of Steps')
plt.show()
```
