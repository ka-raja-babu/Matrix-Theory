```python
#Code by K.A. Raja Babu
#July 15,2021

import matplotlib.pyplot as plt
import numpy as np

x0=np.linspace(-1,0,200)
x1=np.linspace(0,1,200)
y0=(x0*x0)/2+x0+(1/2)
y1=-(x1*x1)/2+x1+(1/2)

plt.plot(x0,y0,label='$x^2$/2+$x$+1/2',color='black')
plt.plot(x1,y1,label='-$x^2$/2+$x$+1/2',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('CDF')
plt.plot(0,0.5,'o')
plt.grid()
plt.legend()
plt.show()
```
